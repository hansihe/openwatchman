```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
sample_endpoint_val = (np.pi/2)-0.1
print(sample_endpoint_val)
sample_points = np.linspace(-sample_endpoint_val, sample_endpoint_val, num=100)
samples = [np.sin(sample_points), np.cos(sample_points)]

samples_points = np.transpose(samples)
projections_x = np.empty(len(samples_points))
for (idx, (x, y)) in enumerate(samples_points):
    n = 1/y
    rx = n*x
    projections_x[idx] = rx
```

```python
plt.scatter(samples[0], samples[1])
plt.scatter(projections_x, np.ones(len(projections_x)))
plt.show()
```

```python
deviation_from_linear = projections_x/sample_points

plt.plot(sample_points, projections_x)
plt.plot(sample_points, sample_points)
plt.plot(sample_points, deviation_from_linear)

func()
```
